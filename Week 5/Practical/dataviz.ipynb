# Matplotlib Intro

```python
# Note: Run this file in google collab
import pandas as pd
import matplotlib.pyplot as plt
# Mount Google Drive if your CSV file is stored there
from google.colab import drive
drive.mount('/content/drive')

# Replace 'your_file_path.csv' with the actual path to your CSV file
file_path = '/content/drive/MyDrive/Colab Notebooks/credit.csv'



# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(file_path, delimiter=' ')

# Display the DataFrame
df.head()
```

```python


column_x = 'Age'
column_y = 'Income'

# Scatter plot
plt.figure(figsize=(4, 3))
plt.scatter(df[column_x], df[column_y])

# Set plot title and labels
plt.title('Scatter Plot')
plt.xlabel(column_x)
plt.ylabel(column_y)

# Show the plot
plt.show()
```

```python


# Specify the columns for the bar chart
x_column = 'Gender'
y_column = 'Income'

# Create a bar chart
plt.figure(figsize=(4, 3))
plt.bar(df[x_column], df[y_column])

# Add labels and title
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title('Bar Chart')

# Show the plot
plt.show()
```

```python


gender_counts = df['Gender'].value_counts()
gender_counts.head()
plt.figure(figsize=(4, 4))
plt.pie(gender_counts, labels=gender_counts.index)
# Add title
plt.title('Distribution of Gender')

# Show the plot
plt.show()
```

```python


ethn_counts = df['Ethnicity'].value_counts()
# ethn_counts.head()
plt.figure(figsize=(4, 4))
plt.pie(ethn_counts, labels=ethn_counts.index)
# Add title
plt.title('Distribution of Ethnicity')

# Show the plot
plt.show()
```

```python


data = {
    'Year': ['2010', '2011', '2012', '2013', '2014'],
     'Income': [20000, 30000, 18000, 30000, 35000],
     'Balance': [100000, 250000, 240000, 130000, 120000  ]
    }
df = pd.DataFrame(data)

# Plot a line graph
plt.figure(figsize=(4, 3))
plt.plot(df['Year'], df['Income'], label='line 1')
plt.plot(df['Year'], df['Balance'], label='line 2')
plt.legend()
# Add labels and title
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Line Graph')

# Show the plot
plt.grid(True)
plt.show()
```
